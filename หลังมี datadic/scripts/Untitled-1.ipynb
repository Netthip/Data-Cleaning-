```python
from pathlib import Path
import os, sys, subprocess

PROJECT = Path(r"C:\Users\netthip\OneDrive\เรียน ป.โท\IS\หลังมี datadic").resolve()
SCRIPTS = PROJECT / "scripts"
CONFIG  = PROJECT / "config"
DATA    = PROJECT / "data"
OUT     = PROJECT / "output"
OUT.mkdir(parents=True, exist_ok=True)
os.chdir(PROJECT)

def run_and_show(cmd_list):
    env = os.environ.copy()
    env["PYTHONIOENCODING"] = "utf-8"
    print("CMD:", " ".join(map(str, cmd_list)))
    res = subprocess.run(
        cmd_list,
        cwd=str(PROJECT),
        env=env,
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True,
        encoding="utf-8",
        errors="replace",
    )
    print("---- STDOUT ----\n", res.stdout)
    print("---- STDERR ----\n", res.stderr)
    print("Exit code:", res.returncode)
    return res

cmd = [
    sys.executable, "-X", "utf8",
    str(SCRIPTS / "build_master.py"),
    "--input_xlsx", str(DATA / "1748328157_5318.xlsx"),
    "--sheet", "data",
    "--config_yml", str(CONFIG / "budget_mapping.yml"),
    "--out_dir", str(OUT),
]
res = run_and_show(cmd)

```

```python
import pandas as pd, os, numpy as np

OUT = r"C:\Users\netthip\OneDrive\เรียน ป.โท\IS\หลังมี datadic\output"
md  = pd.read_excel(fr"{OUT}\MasterData.xlsx")

print("rows:", len(md))
print("cols:", list(md.columns))
print("sum p_total_bud:", md["p_total_bud"].fillna(0).sum())
print("unknown % by amount:",
      md.loc[md["budget_group"]=="ไม่ทราบหมวด","p_total_bud"].fillna(0).sum() /
      md["p_total_bud"].fillna(0).sum() * 100)
md.sample(5)
```

```python
import pandas as pd

pvt1 = md.pivot_table(index="min_name", columns="budget_group",
                      values="p_total_bud", aggfunc="sum", fill_value=0)

pvt2 = md.pivot_table(index="min_name", columns="expenditure_type",
                      values="p_total_bud", aggfunc="sum", fill_value=0)

pvt3 = md.pivot_table(index=["plan_name","output_name","act_name","budget_group"],
                      values="p_total_bud", aggfunc="sum", fill_value=0).reset_index()

OUT = r"C:\Users\netthip\OneDrive\เรียน ป.โท\IS\หลังมี datadic\output"
with pd.ExcelWriter(fr"{OUT}\PivotQuickCheck.xlsx", engine="openpyxl") as w:
    pvt1.to_excel(w, sheet_name="Min x BudgetGroup")
    pvt2.to_excel(w, sheet_name="Min x Type")
    pvt3.to_excel(w, sheet_name="Plan-Output-Act")
print("saved:", fr"{OUT}\PivotQuickCheck.xlsx")
```

```python
# ลองเขียนเเฉพาะหน่วยงานที่ทำจริง ๆ 
focus = md[md["agc_name"].str.contains("สถาบันวัคซีน|สถาบันวิจัย", na=False)]
with pd.ExcelWriter(fr"{OUT}\Focus_Vaccine_Research.xlsx", engine="openpyxl") as w:
    focus.to_excel(w, index=False, sheet_name="detail")
    focus.pivot_table(index="agc_name", columns="budget_group",
                      values="p_total_bud", aggfunc="sum", fill_value=0).to_excel(w, sheet_name="pivot")
print("saved:", fr"{OUT}\Focus_Vaccine_Research.xlsx")
```
